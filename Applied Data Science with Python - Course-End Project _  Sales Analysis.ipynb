```python
#Set up different packages needed for 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load your CSV data into a DataFrame
df = pd.read_excel('/Users/sheiljasheth/Downloads/1688639662_ausapparalsales4thqrt2020.xlsx')
```

```python
# Data Wrangling
# Check for missing or incorrect data
missing_data = df.isna()
incorrect_data = df.notna()
```

```python
df['Sales_Standardized'] = (df['Sales'] - df['Sales'].mean()) / df['Sales'].std()
```

```python
sales_stats = df[['Group', 'Sales']].groupby('Group').describe()
state_stats = df[['State', 'Sales']].groupby('State').describe()
```

## Descriptive statistical analysis

### Sales Data 

```python
sales_stats = df['Sales'].describe()
sales_mean = df['Sales'].mean()
sales_median = df['Sales'].median()
sales_mode = df['Sales'].mode().values[0]
sales_std = df['Sales'].std()

# Display the results
print("Descriptive Statistics for Sales:")
print(sales_stats)
print(f"Mean: {sales_mean}")
print(f"Median: {sales_median}")
print(f"Mode: {sales_mode}")
print(f"Standard Deviation: {sales_std}")
```

### Unit Data 

```python
unit_stats = df['Unit'].describe()
unit_mean = df['Unit'].mean()
unit_median = df['Unit'].median()
unit_mode = df['Unit'].mode().values[0]
unit_std = df['Unit'].std()

# Display the results
print("\nDescriptive Statistics for Unit:")
print(unit_stats)
print(f"Mean: {unit_mean}")
print(f"Median: {unit_median}")
print(f"Mode: {unit_mode}")
print(f"Standard Deviation: {unit_std}")
```

## Highest and Lowest Sales by Group

```python
group_sales = df.groupby('Group')['Sales'].sum()
highest_selling_group = group_sales.idxmax()
lowest_selling_group = group_sales.idxmin()

#Print results 
print(f"The group generating the highest sales is: {highest_selling_group}")
print(f"The group generating the lowest sales is: {lowest_selling_group}")
```

## Highest and Lowest Sales by State

```python
state_sales = df.groupby('State')['Sales'].sum()
highest_selling_state = state_sales.idxmax()
lowest_selling_state = state_sales.idxmin()

#Print results 
print(f"The state generating the highest sales is: {highest_selling_state}")
print(f"The state generating the lowest sales is: {lowest_selling_state}")
```

## Sales over Time

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Sales'], label='Sales', color='b')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

## Sales by State

```python
plt.figure(figsize=(15, 8))
sns.barplot(x='State', y='Sales', data=df)
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

## State-wise Sales Analysis by Group

```python
plt.figure(figsize=(15, 8))
sns.barplot(x='State', y='Sales', hue='Group', data=df)
plt.title('State-wise Sales Analysis by Group')
plt.show()
```

## Group-wise sales analysis across different states

```python
plt.figure(figsize=(15, 8))
sns.barplot(x='Group', y='Sales', hue='State', data=df)
plt.title('Group-wise Sales Analysis by State')
plt.show()
```

## Time-of-the-day analysis 

```python
time_sales = df.groupby('Time')['Sales'].sum()
time_sales.plot()
plt.title('Time-of-the-Day Sales Analysis')
plt.show()
```

# Daily, Weekly, and Monthly Visualizations

## Daily Sales

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
daily_sales = df['Sales'].resample('D').sum()
daily_sales.plot()
plt.title('Daily Sales')
```

## Weekly Sales

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 2)
weekly_sales = df['Sales'].resample('W').sum()
weekly_sales.plot()
plt.title('Weekly Sales')
```

## Monthly Sales

```python
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 3)
monthly_sales = df['Sales'].resample('M').sum()
monthly_sales.plot()
plt.title('Monthly Sales')
```

## Box Plot Sales by Group 

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='Group', y='Sales', data=df)
plt.title('Box Plot for Sales by Group')
plt.xlabel('Group')
plt.ylabel('Sales')
plt.show()
```

## Sales Distribution

```python
plt.figure(figsize=(15, 8))
sns.histplot(data=df, x='Sales', kde=True)
plt.title('Distribution Plot for Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

## Sales by Unit

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Sales', hue='Unit', kde=True)
plt.title('Distribution Plot for Sales by Unit')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.legend(title='Unit')
plt.show()
```

## Units by Group 

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Unit', hue='Group', kde=True)
plt.title('Distribution Plot for Unit by Group')
plt.xlabel('Group')
plt.ylabel('Unit')
plt.legend(title='Group')
plt.show()
```
