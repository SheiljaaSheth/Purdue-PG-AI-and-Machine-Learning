```python
# product catalog
products = [
    {"product_id": 1, "name": "Boots", "category": "Footwear", "price": 50},
    {"product_id": 2, "name": "Coats", "category": "Clothing", "price": 100},
    {"product_id": 3, "name": "Jackets", "category": "Clothing", "price": 80},
    {"product_id": 4, "name": "Caps", "category": "Accessories", "price": 20},
]

# user and admin credentials
user_credentials = {"username": "user", "password": "password123"}
admin_credentials = {"username": "admin", "password": "admin_password"}

# user cart
user_cart = []

# admin session
admin_session = None

# Function to display the welcome message
def display_welcome_message():
    print("Welcome to the Demo Marketplace")

# Function for user login
def user_login():
    global admin_session
    username = input("Enter your username: ")
    password = input("Enter your password: ")
    
    if username == user_credentials["username"] and password == user_credentials["password"]:
        print("User login successful.")
        return True
    elif username == admin_credentials["username"] and password == admin_credentials["password"]:
        print("Admin login successful.")
        admin_session = True
        return True
    else:
        print("Login failed. Incorrect username or password.")
        return False

# Function to display the product catalog
def display_product_catalog():
    print("Product Catalog:")
    for product in products:
        print(f"Product ID: {product['product_id']}, Name: {product['name']}, Category: {product['category']}, Price: Rs. {product['price']}")

# Function to add items to the cart
def add_to_cart():
    product_id = int(input("Enter the Product ID to add to cart: "))
    quantity = int(input("Enter the quantity: "))
    
    product = next((p for p in products if p["product_id"] == product_id), None)
    
    if product:
        user_cart.append({"product_id": product["product_id"], "name": product["name"], "price": product["price"], "quantity": quantity})
        print(f"{quantity} {product['name']} added to cart.")
    else:
        print("Product not found in catalog.")

# Function to remove items from the cart
def remove_from_cart():
    product_id = int(input("Enter the Product ID to remove from cart: "))
    
    for item in user_cart:
        if item["product_id"] == product_id:
            user_cart.remove(item)
            print(f"{item['name']} removed from cart.")
            return
    print("Product not found in the cart.")

# Function to display payment options and checkout
def checkout():
    print("Payment Options:")
    print("1. Net banking")
    print("2. PayPal")
    print("3. UPI")
    print("4. Debit/Credit")
    
    option = int(input("Select a payment option: "))
    
    if option in [1, 2, 3, 4]:
        print("Your order is successfully placed.")
    else:
        print("Invalid payment option selected.")

# Main program loop
while True:
    display_welcome_message()
    authenticated = user_login()
    
    if authenticated:
        display_product_catalog()
        if admin_session:
            # Admin-specific actions (add, modify, remove products and categories)
            pass
        else:
            # User-specific actions (view cart, add/remove items, checkout)
            print("User Actions:")
            print("1. View Cart")
            print("2. Add to Cart")
            print("3. Remove from Cart")
            print("4. Checkout")
            
            choice = int(input("Select an action: "))
            
            if choice == 1:
                print("Cart Contents:")
                for item in user_cart:
                    print(f"Product: {item['name']}, Price: Rs. {item['price']}, Quantity: {item['quantity']}")
            elif choice == 2:
                add_to_cart()
            elif choice == 3:
                remove_from_cart()
            elif choice == 4:
                checkout()
            else:
                print("Invalid choice.")
    
    another_transaction = input("Do you want to perform another transaction? (yes/no): ")
    if another_transaction.lower() != "yes":
        break
```
